```python
### make clipped videos of th behaviour for training DLC

import os 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
```

```python
# load in the videos for the mouse:
path = r"Z:\projects\sequence_squad\revision_data\organised_data\animals\seq052_implant1\\"

```

```python
for recording in os.listdir(path):
    video_path = os.path.join(path, recording) + r'\video\videos\\'
    
    print(recording)

    for file in os.listdir(video_path):
        if '.avi' in file and 'BACK' in file:
            raw_cam_path = os.path.join(video_path, file)
            video_file = file
        elif '.csv' in file and 'BACK' in file:
            Camera_dataframe = pd.read_csv(os.path.join(video_path, file))
            
    trigger_changes = np.where(np.diff(Camera_dataframe['Trigger State'].values) == 1)[0]
    trigger_onset_times = Camera_dataframe['Time Stamps'].values[trigger_changes]
        
    sleep_trigger_inds = np.where((np.diff(trigger_onset_times) > 28.5) & (np.diff(trigger_onset_times) < 31.5))


    consecutive_chunks = []
    start = None
    for i in range(len(sleep_trigger_inds[0]) - 1):
        if sleep_trigger_inds[0][i + 1] - sleep_trigger_inds[0][i] == 1:
            if start is None:
                start = sleep_trigger_inds[0][i]
        else:
            if start is not None:
                end = sleep_trigger_inds[0][i]
                consecutive_chunks.append((start, end))
                start = None
    if start is not None:
        consecutive_chunks.append((start, sleep_trigger_inds[0][-1]))
        
    consecutive_chunks[1]

    sleep1_end_trigger = consecutive_chunks[0][-1]
    # sleep1_end_trigger = trigger_changes[0]
    sleep2_start_trigger = consecutive_chunks[-1][0]

    cut1_s = trigger_onset_times[sleep1_end_trigger]
    cut2_s = trigger_onset_times[sleep2_start_trigger]

    out_path = r"Z:\projects\sequence_squad\revision_data\emmett_revisions\DLC_video_dump\videos_for_training_DLC\\"

    # Specify your times in seconds (floats allowed) or "HH:MM:SS" strings
    start_time = cut1_s         # 15 seconds in
    end_time   = cut2_s      # 42.5 seconds

    ffmpeg_extract_subclip(raw_cam_path,start_time ,end_time,out_path+video_file)
        
    print(f"Video saved to {out_path+video_file}")
                

        
```
